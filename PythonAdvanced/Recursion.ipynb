# Understanding Recursion

- We call the same function within itself

```python
'''
Factorial of a number
5! = 5 * 4!
4! = 4 * 3! and so on
'''

def fact_func(num1):
    if num1 == 0 or num1 == 1:
        return 1
    return num1 * fact_func(num1-1)

print(fact_func(2))
```
