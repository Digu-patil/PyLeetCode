```python
# Lets write a small code to add 2 numbers

code = b"def add(a,b):return a+b"
```

```python
'''
Tokenization
1. Source code (.py) is converted to tokens
2. Here we check of lexical error.
'''

import tokenize
from io import BytesIO

#Lets convert th e code to bytes
byte_lines = BytesIO(code).readline

# print(byte_lines)

for tok in tokenize.tokenize(byte_lines):
    print(tok)
```

```python
'''
Parsing (AST generation)
1. Here we check if the syntax is correct or not
2. We now have tokens, we just need to know if the tokens make sense grammatically
'''

import ast
tree = ast.parse(code)
print(ast.dump(tree, indent=1))
```

```python
'''
Bytecode generation
**note that this bytecode is in human readable format, gnerally its different

This byte code is then read by the Python Virtual Machine
'''

import dis
dis.dis("def add(a,b):return a+b")
```
